### Where to take this next
These are the natural extensions that bring it closer to a real Depop-scale system:
A vector database is the obvious next layer — embed your full catalogue offline, store the vectors in OpenSearch, FAISS, or Pinecone, and at query time embed the query image and do a nearest-neighbour search. That's an actual retrieval system rather than just an embedding service.
Swap ResNet for CLIP once the pipeline works. CLIP gives you text-to-image search ("red vintage jacket" → matching items), which is much closer to how Depop search actually works.
For MLOps maturity, set up GitHub Actions to build and push the image on every commit to main, add latency and error-rate monitoring via CloudWatch, and define your SageMaker resources in Terraform or CDK so the whole thing is reproducible. That trio — CI/CD, monitoring, IaC — is exactly the "MLOps environment" the JD calls out.

### Step 5: Deploy to SageMaker as a real-time endpoint
SageMaker has three serving flavours; the JD says "real-time" so we want the real-time endpoint (always-on instance, low latency, HTTPS API). The alternatives — Serverless Inference (scales to zero, cold starts) and Async Inference (for long jobs) — are worth knowing about but not what you want here.
